```python
import pandas as pd
import rqdatac
```

```python
def get_profolio_factor(profolio_name,start_date,end_date):
    '''这是用econ get factor 向前填充得到的因子值
    输入： factor name
    输出：
    index为发行日期‘date’的数据
    column是向前填充的标准化之后的因子值 注意：一定将因子值的名称改为‘value’

    '''
    factor_df = rqdatac.get_price(profolio_name,start_date,end_date).reset_index()
    factor_df['value'] = factor_df['close'].pct_change()
    factor_df = factor_df[['date', 'value']]
    factor_df = factor_df.dropna()
    factor_df = factor_df.set_index('date')

    return factor_df
```
